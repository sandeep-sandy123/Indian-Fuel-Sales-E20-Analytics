### Convert the date column if necessary

### Revenue by State

#### Which state generated the highest revenue?

```python
state_revenue = df.groupby("State")["Revenue"].sum().sort_values(ascending=False)

print(state_revenue)
```

#### Revenue by Fuel Type

```python
fuel_revenue = df.groupby("Fuel_Type")["Revenue"].sum()

print(fuel_revenue)
```

#### Revenue by Customer Type

```python
customer_revenue = df.groupby("Customer_Type")["Revenue"].sum()

print(customer_revenue)
```

#### Monthly Revenue Trend

```python
df["Month"] = df["Date"].dt.month_name()

monthly_revenue = df.groupby("Month")["Revenue"].sum()

print(monthly_revenue)
```

#### Correlation Analysis

##### How are Quantity, Price, and Revenue related?

```python
correlation = df[["Quantity_Litres","Price_Per_Litre","Revenue"]].corr()

print(correlation)
```

### Outlier Detection (IQR Method)

```python
Q1 = df["Revenue"].quantile(0.25)
Q3 = df["Revenue"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df["Revenue"] < lower) | (df["Revenue"] > upper)]

print(outliers)
```
